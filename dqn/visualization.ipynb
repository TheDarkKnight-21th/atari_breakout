```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
score = pd.read_csv("score_record.csv")
loss = pd.read_csv("loss.csv")
```

```python
score
```

```python

plt.figure(figsize=(10,5))
plt.title("Average score")
plt.style.use('ggplot')
plt.plot(score.step , score.eval_score , color = "blue")
plt.xlabel("step")
plt.ylabel("score")
plt.show()
```

```python

plt.figure(figsize=(10,5))
plt.title("loss")
plt.style.use('ggplot')
plt.plot(loss.step , loss.loss , color = "orange")
plt.xlabel("step")
plt.ylabel("loss")
plt.show()
```
